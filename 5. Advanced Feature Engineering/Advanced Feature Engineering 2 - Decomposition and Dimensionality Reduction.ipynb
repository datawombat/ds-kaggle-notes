One should already know that MF (Matrix Factorization) is done for dimensionality reduction.

A way to use it as a feature engineering method can be:
- To reduce dimensionality of a whole dataset
- To join large numbers of features
- To reduce dimensionality of only some columns
- To provide additional diversity (because if it's reduced, then it's another representation) for ENSEMBLES.

## NMF

This is how the MF algorithm is called: Non-negative matrix factorization or approximation (NMF or NNMF). So it doesn't receive negative values.

Therefore, a standartized matrix cannot be the input.

## Joining large numbers of features

Therefore, using MF gives "new features" that will finally serve as input.

Common types of features from text data:
- Vanilla BOW
- BOW + TF-IDF
- BOX (bigrams)

One can put all of them in a single matrix after DIMENSIONALITY REDUCTION.

![](images/dimensionality_red.png)

Another example:

Preprocesing data like with LOG(X) can be done before matrix factorization.

Then, instead of choosing the preprocessed or the original. They can both be used together.

**Joint features = NMF(X), NMF(log(X+1))**
