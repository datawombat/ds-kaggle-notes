Ensemble methods are ML methods for combining several base models in order to produce one optimal predictive model.

## Why Ensembling?

The **bias-variance tradeoff** states that for a single machine learning model, bias and variance cannot be reduced simultaneously.

For example, regularization reduces the variance, causing underfitting.

But, ENSEMBLE METHODS methods can reduce both bias and variance so that a model generalizes better.


## Most Common Types of Ensembling

- Averaging (or blending) and Weighted Averaging
- Conditional Averaging
- Bagging
- **Boosting** (super popular)
- Stacking
- StackNet (library)

There are others but those seem to be the most effective as well.
